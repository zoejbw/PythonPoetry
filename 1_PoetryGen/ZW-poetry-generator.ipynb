```python
stanza_count = 5
for i in range(stanza_count):
    brand = random.choice(username) + random.choice(filler) + random.choice(username)
    description = "[" + random.choice(imageType) + random.choice(subject) + random.choice(verb) + random.choice(thing) + "]"
    callToAction = random.choice(links)
    likenum = random.randint(2, 999)
    
    ##changes Likes to Views if description starts with "looping video"
    if description.startswith("[l"):
        liketype = " Views"
        likenum = likenum*random.randint(2, 5)
    else:
        liketype = " Likes"
    
    ##adds commas if view count is over 999
    likenum = str(likenum)
    if len(likenum) > 3:
        likenum = str(likenum[:len(likenum)-3]) + "," + str(likenum[len(likenum)-3:]) 
    
    ##creates caption until caption is long enough
    caption = random.choice(captionA) + random.choice(captionB) + random.choice(captionC)
    line = brand + ": " + caption   
    while len(line) < 91:
        line = line + random.choice(captionA) + random.choice(captionB) + random.choice(captionC)  
       
    ##finds space to make line break
    i = 45
    while line[i] != " ":
        i = i-1
    
    lineA = line[:i]
    lineB = line[i+1:i+35]
    
    
    commentnum = str(random.randint(2, 100))
    
    print()
    print(brand)
    print("Sponsored")
    print(description)
    print(callToAction)
    print(likenum + liketype)
    print(lineA)
    print(lineB + "... more")
    print("View " + commentnum + " comments")
```
