```python
clueless = open("plaintext/Clueless.txt").read()
emma = open("plaintext/Emma.txt").read()
```

```python
#Gets rid of weird screenplay formatting
clueless = clueless.replace("\n\n","\n")

#Changes Clueless names to their Emma equivalents
clueless = clueless.replace("Cher","Emma")
clueless = clueless.replace("Tai","Harriet")
clueless = clueless.replace("Josh","Mr. Knightly")
clueless = clueless.replace("Elton","Mr. Elton")
clueless = clueless.replace("Christian","Frank")
clueless = clueless.replace("Travis","Robert")
clueless = clueless.replace("Dionne","Mrs. Weston")
clueless = clueless.replace("Horowitz","Woodhouse")
clueless = clueless.replace("Mel","Mr. Woodhouse")
clueless = clueless.replace("Murray","Mr. Weston")

#Changes Clueless names IN DIALOGUE to their Emma equivalents
clueless = clueless.replace("CHER","Emma:")
clueless = clueless.replace("TAI","Harriet:")
clueless = clueless.replace("JOSH","Mr. Knightly:")
clueless = clueless.replace("ELTON","Mr. Elton:")
clueless = clueless.replace("CHRISTIAN","Frank:")
clueless = clueless.replace("TRAVIS","Robert:")
clueless = clueless.replace("DIONNE","Mrs. Weston:")
clueless = clueless.replace("MEL","Mr. Woodhouse:")
clueless = clueless.replace("MURRAY","Mr. Weston:")

#Removes title info 
clueless = clueless[3193:]
emma = emma[620:]
```

```python
emma = emma.split()
clueless = clueless.split(" ")
newtxt = ""

#Choose starting text
isEmma = True
#Choose where Emma starts
eNum = 1
#Choose where Clueless Starts
cNum = 0
#Choose how many words to search for match through
allowance = 20
#Choose length of text
while len(newtxt) < 5000:
    #Resets word values to avoid index error
    if eNum >= len(emma):
        eNum = 0
    elif cNum >= len(clueless):
        cNum = 0 
    #if on Emma, iterate through words in Clueless looking for a match
    elif isEmma:
        for i in range(cNum, cNum + allowance):
            if (emma[eNum].lower() == clueless[i].lower()):  
                isEmma = False
                newtxt = newtxt + " " + clueless[i]
                cNum = i + 1
                break
        #if no match is found, move on to next word
        else:
            newtxt = newtxt + " " + emma[eNum]
            eNum = eNum + 1
    #if on Clueless, iterate through words in Emma looking for a match
    else:
        for i in range(eNum, eNum + allowance):
            if (clueless[cNum].lower() == emma[i].lower()):          
                isEmma = True
                newtxt = newtxt + " " + emma[i]
                eNum = i + 1
                break
        #if no match is found, move on to next word
        else:
            newtxt = newtxt + " " + clueless[cNum]
            cNum = cNum + 1
            
print(newtxt) 
```
