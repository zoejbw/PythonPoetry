```python
rj = open("plaintext/RandJ.txt").read()
west = open("plaintext/west.txt").read()
```

```python
#Changes WestSide names IN DIALOGUE to their RJ equivalents
west = west.replace("TONY","Rom.")
west = west.replace("MARIA","Jul.")
west = west.replace("ANITA","Nurse.")
west = west.replace("RIFF","Mer.")
west = west.replace("BERNARDO","Tyb.")
west = west.replace("CHINO","Par.")
west = west.replace("JETS","Montagues.")
west = west.replace("SHARKS","Capulets.")
west = west.replace("KRUPKE","Prince.")
west = west.replace("BABY JOHN","Ben.")

#Changes WestSide names to their RJ equivalents
west = west.replace("Tony","Romeo")
west = west.replace("Maria","Juliet")
west = west.replace("Anita","Nurse")
west = west.replace("Riff","Mercutio")
west = west.replace("Bernardo","Tybalt")
west = west.replace("Chino","Paris")
west = west.replace("Jet","Montague")
west = west.replace("Shark","Capulet")
west = west.replace("Officer","Prince")
west = west.replace("Krupke","Escalus")
west = west.replace("Baby John","Benvolio")

#Removes Title Info
west = west[2358:]
rj = rj[2850:]
```

```python
west = west.split(" ")
rj = rj.split(" ")
newtxt = ""

#Choose starting text
isR = True

#Choose where rj starts
rNum = 50

#Choose where west Starts
wNum = 0

#Choose how many words to search for match through
allowance = 20

#Choose length of text
while len(newtxt) < 5000:
    #Resets word values to avoid index error
    if rNum >= len(rj):
        rNum = 0
    elif wNum >= len(west):
        wNum = 0 
    #if on RJ, iterate through words in west looking for a match
    elif isR:
        for i in range(wNum, wNum + allowance):
            if (rj[rNum].lower() == west[i].lower()):  
                isR = False
                newtxt = newtxt + " " + west[i]
                wNum = i + 1
                break
        #if no match is found, move on to next word
        else:
            newtxt = newtxt + " " + rj[rNum]
            rNum = rNum + 1
    #if on west, iterate through words in RJ looking for a match
    else:
        for i in range(rNum, rNum + allowance):
            if (west[wNum].lower() == rj[i].lower()):          
                isR = True
                newtxt = newtxt + " " + rj[i]
                rNum = i + 1
                break
        #if no match is found, move on to next word
        else:
            newtxt = newtxt + " " + west[wNum]
            wNum = wNum + 1
            
print(newtxt) 
```
